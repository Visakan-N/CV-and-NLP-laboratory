# Image Segmentation

```python
import matplotlib.pyplot as plt 
```

```python
image = plt.imread('butterfly.jpg')
plt.imshow(image)
```

```python
from skimage.color import rgb2gray
gray = rgb2gray(image)
plt.imshow(gray, cmap='gray')
```

<b> We take the mean of the image as the threshold value for segmentation

```python
segmented = gray.reshape(gray.shape[0] * gray.shape[1]) # converting image to a vector
threshold = segmented.mean() 
```

```python
for i in range(segmented.shape[0]):
    if segmented[i] > threshold:
        segmented[i] = 1
    else:
        segmented[i] = 0

segmented = segmented.reshape(gray.shape[0], gray.shape[1])
plt.imshow(segmented, cmap='gray')
```
