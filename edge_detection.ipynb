# Edge Detection

```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('butterfly.jpg')
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
blurred = cv2.GaussianBlur(gray, (3, 3), 0)
```

```python
edges = cv2.Canny(blurred, 50, 150)
```

```python
plt.imshow(edges, cmap='gray')
```
