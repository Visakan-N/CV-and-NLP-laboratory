# Pattern Recognition
Template matching - using a smaller image (template) to located its position in another larger image.

```python
import cv2
import matplotlib.pyplot as plt
```

```python
template = cv2.imread('football.jpg', 0) # 0 indicates to load image in grayscale
image = cv2.imread('goal.jpg', 0)
```

```python
result = cv2.matchTemplate(image, template, cv2.TM_CCOEFF_NORMED)
```

```python
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result) # coordinates of best match
```

```python
top_left = max_loc
bottom_right = (top_left[0] + template.shape[1], top_left[1] + template.shape[0])
```

```python
matched_image = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)
cv2.rectangle(matched_image, top_left, bottom_right, (0, 0, 255), 2) # drawing a rectangle around the matched image
```

```python
plt.imshow(cv2.cvtColor(matched_image, cv2.COLOR_BGR2RGB))
```
